```python
import pandas as pd
# Ruta al archivo subido
file_path = "D:\\Users\\OneDrive\\Escritorio\\DA & BA\\Work\\Tisagar\\Bitso\\Bitso 1er semana julio.csv"

# Leer el archivo como texto y eliminar las comillas dobles
with open(file_path, 'r', encoding='utf-8') as file:
    content = file.read()

# Reemplazar las comillas dobles
content = content.replace('"', '')

# Guardar el contenido limpio en un nuevo archivo temporal
temp_file_path = "D:\\Users\\OneDrive\\Escritorio\\DA & BA\\Work\\Tisagar\\Bitso\\Bitso 1er semana julio.csv"
with open(temp_file_path, 'w', encoding='utf-8') as file:
    file.write(content)

# Leer el archivo limpio con pandas
df = pd.read_csv(temp_file_path, delimiter=';')
```

```python
columnas_delete = ['amount','fee']
df = df.drop(columns=columnas_delete)
```

```python
# Intentar corregir los formatos si es necesario (ejemplo para formatos sin segundos)
df['datetime'] = df['datetime'].str.replace(r'(\d{1,2}/\d{1,2}/\d{4} \d{2}:\d{2})$', r'\1:00', regex=True)
```

```python
# Intentar convertir la columna de fecha sin especificar el formato
df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')
```

```python
# Extraer solo la fecha (sin la hora)
df['datetime'] = df['datetime'].dt.strftime('%m/%d/%Y')
```

```python
df_compra = df[df['type'] == 'buy']
df_venta = df[df['type'] == 'sell']
```

```python
df_c = df_compra.sort_values(by='datetime')
df_v = df_venta.sort_values(by='datetime')
```

```python
df_c['Exchange compra'] = 'Bitso - Usdt'
df_c['Banco envio'] = 'Lemon'

df_v['Exchange Venta'] = 'Bitso - Usdt'
df_v['Banco recibo'] = 'Lemon'
```

```python
columnas_ventas = ['datetime', 'Exchange Venta', 'value', 'rate', 'total', 'Banco recibo', 'timestamp','type']
columnas_compra = ['datetime','Exchange compra', 'value', 'rate', 'total', 'Banco envio', 'timestamp','type']
```

```python
df_v = df_v[columnas_ventas]
df_c = df_c[columnas_compra]
```
